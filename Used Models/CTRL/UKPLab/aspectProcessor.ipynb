```python
import pandas as pd
import numpy as np
import csv
from pathlib import Path
import spacy
import numpy as np
import pandas as pd
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = stopwords.words("english")
import platform
import argparse
from tqdm import tqdm
import enchant
dnc = enchant.Dict("en_US")
```

```python
#file = pd.read_csv('ExtractedPro.txt', sep=':::')
file = pd.read_csv('ExtractedCon.txt', sep=':::')
file.head()
```

```python
id = file['id'].tolist()
topic = file['topic'].tolist()
text = file['text'].tolist()
```

```python
import string
```

```python
def typecheck(s):
  de = False
  if s.isdigit():
    de = True
  elif s.startswith('-') and s[1:].isdigit():
    de = True
  for c in s:
    if c in string.punctuation:
      de = True
  return de
```

```python
a = []
b = []
c = []
for x in tqdm(range(len(topic))):
  if dnc.check(str(topic[x])):
    if typecheck(topic[x]) == False:
       a.append(id[x])
       b.append(topic[x])
       c.append(text[x])
```

```python
#f = open("FilteredPro.txt", 'a')
f = open("FilteredCon.txt", 'a')
f.write("id:::topic:::text")
for x in tqdm(range(len(b))):
   f.write("\n")
   f.write(str(a[x]))
   f.write(":::")
   f.write(str(b[x]))
   f.write(":::")
   f.write(str(c[x]))
f.close()
```
